##1.2, Mục tiêu:

Mục tiêu của báo cáo này là đánh giá xem loại nội dung nào (hình ảnh hay video) mang lại hiệu quả chuyển đổi cao hơn khi chạy quảng cáo trên Facebook.
Phương pháp triển khai: Thiết lập thí nghiệm A/B testing để đánh giá.

Các chỉ số chính:

1, **Tỷ lệ chuyển đổi:**

$${\text{CR} = \frac{\text{Tổng số tin nhắn khách hàng mới được chuyển đổi}}{\text{Tổng số người tiếp cận}}}$$

Ý nghĩa: Tỷ lệ này cho biết trong số những người tiếp cận được quảng cáo, bao nhiêu khách hàng mới thực hiện hành động nhắn tin.

2, **Tỷ lệ nhấp chuột:**

$${\text{CTR} = \frac{\text{Tổng số lượt nhấp chuột}}{\text{Tổng số lần hiển thị}}}$$

Ý nghĩa: Cho biết mức độ hấp dẫn của quảng cáo. Quảng cáo video thường có khả năng thu hút sự chú ý của người xem tốt hơn, nhưng tỷ lệ nhấp chuột sẽ giúp so sánh sự hiệu quả giữa hình ảnh và video trong việc kích thích người xem hành động.

3, **Tỷ lệ tương tác bài viết:**

$${\text{ERpost} = \frac{\text{Tổng số lượt cảm xúc, bình luận, chia sẻ, lưu bài viết}}{\text{Tổng số lần hiển thị}}}$$

Ý nghĩa: Đo lường mức độ tương tác của người dùng với quảng cáo, bao gồm lượt thích, bình luận và chia sẻ.

4, **Tỷ lệ chuyển đổi tương tác page:**

$${\text{ERpage} = \frac{\text{Tổng số lượt theo dõi, thích page}}{\text{Tổng số lần hiển thị}}}$$

Ý nghĩa: Đo lường mức độ tương tác của người dùng với trang page thông qua quảng cáo.

#4, Kiểm định thống kê và Phân tích kết quả:

Tại sao cần kiểm định thống kê:
Kiểm định thống kê để biết 2 tỷ lệ ta đo được trong 2 mẫu có nằm chung 1 phân phối thống kê hay không. Nếu có, mặc dù số liệu có chút khác biệt nhưng tính chất chúng như nhau, đều nằm trong 1 tổng thể có cùng trung bình. Nếu không, sự khác biệt của nó có ý nghĩa thống kê nghĩa là 2 số liệu đó nằm trong 2 phân phối thống kê của 2 tổng thể có trung bình khác nhau.

Có thể tham khảo hình dưới:

image.png


```python
# Viết hàm kiểm định thống kê z-test
import statsmodels.stats.proportion as smp

def kiemdinh(x1,n1,x2,n2):

  # Giả thuyết:
  print('Giả thuyết:\nHo: p1=p2\nHa: p1<>p2\n')

  # Kiểm định Z-test
  stat, p_value = smp.proportions_ztest(count=[x1, x2], nobs=[n1, n2], alternative='two-sided')

  # Kết quả
  print(f"Z-statistic: {stat:.4f}")
  print(f"P-value: {p_value:.4f}")

  # Đánh giá
  alpha = 0.05  # Mức ý nghĩa
  if p_value < alpha:
      print("Kết luận: Bác bỏ giả thuyết H0, tỷ lệ chuyển đổi p1 khác p2")
  else:
       print("Kết luận: Chưa có đủ bằng chứng để bác bỏ giả thuyết H0, tỷ lệ chuyển đổi là như nhau")
```

##4.1, So sánh tỷ lệ chuyển đổi tin nhắn khách hàng mới:

- Nhóm A: 0.37%
- Nhóm B: 0.26%

```python
kiemdinh(71,19144,87,33054)
```

Từ kiểm định trên, ta thấy nhóm A và B có sự khác biệt đáng kể trong thống kê. Cụ thể, tỷ lệ chuyển đổi tin nhắn khách hàng mới của nhóm A (hình ảnh) cao hơn nhóm B (video).

Kết luận: Việc dùng hình ảnh để chạy quảng cáo tăng tỷ lệ chuyển đổi ra tin nhắn khách hàng mới nhiều hơn video.

##4.2, So sánh tỷ lệ nhấp chuột:


- Nhóm A: 11.63%
- Nhóm B: 7.3 %

```python
kiemdinh(4085,35111,3220,44104)
```

Từ kiểm định trên, ta thấy nhóm A và B có sự khác biệt đáng kể trong thống kê. Cụ thể Nhóm A có tỷ lệ nhấp chuột cao hơn đáng kể so với nhóm B.

Kết luận: Việc dùng hình ảnh thu hút người xem nhấp vào quảng cáo nhiều hơn video.

##4.3, So sánh tỷ lệ tương tác bài viết:


- Nhóm A: 0.36%
- Nhóm B: 0.54%

```python
kiemdinh(128,35111,236,44104)
```

Nhóm B có tỷ lệ tương tác bài viết cao hơn so với Nhóm A (0.54% > 0.36%).

Kết luận: Dùng video có mức độ thu hút và tương tác từ người dùng tốt hơn hình ảnh.

##4.4, So sánh tỷ lệ chuyển đổi tương tác page:


- Nhóm A: 0.17%
- Nhóm B: 0.44%

```python
kiemdinh(58,35111,193,44104)
```

Nhóm B có tỷ lệ chuyển đổi tương tác trên page cao hơn đáng kể (0.44% > 0.17%) so với Nhóm A.

Kết luận: Việc dùng video hiệu quả hơn trong việc chuyển đổi người dùng từ các hoạt động trên bài viết (post engagement) thành hành động có sự cam kết lâu dài với shop (like, follow page...)
